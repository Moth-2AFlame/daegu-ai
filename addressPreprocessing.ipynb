설치 연도 2019/2020/2021로 분리

```python
cctv2019 = cctv_df[cctv_df['설치연도'] <= 2019.0]
cctv2020 = cctv_df[cctv_df['설치연도'] <= 2020.0]
cctv2021 = cctv_df[cctv_df['설치연도'] <= 2021.0]

cctv2019.drop(columns = ['설치연도'],inplace = True)
cctv2020.drop(columns = ['설치연도'],inplace = True)
cctv2021.drop(columns = ['설치연도'],inplace = True)
```

```python
cctv2019 = cctv2019.groupby(['시군구']).agg(['count','mean']).reset_index()
cctv2020 = cctv2020.groupby(['시군구']).agg(['count','mean']).reset_index()
cctv2021 = cctv2021.groupby(['시군구']).agg(['count','mean']).reset_index()
```
