## 보안등

```python
light = pd.read_csv("/content/drive/MyDrive/data/dacon/external_open/대구 보안등 정보.csv",encoding = 'cp949')
```

```python
light
```

```python
light_df = light[['설치개수','소재지지번주소']]
```

```python
light_df
```

```python
import re
p = r'(\S+) (\S+) (\S+) (\S+)'

light_df[['도시','구','동','번지']] = light_df['소재지지번주소'].str.extract(p)
```

```python
light_df
```

```python
light_df.drop(columns = ['소재지지번주소','번지'],inplace = True)
```

```python
light_df
```

```python
light_df = light_df.groupby(['도시','구','동']).sum().reset_index()
```

```python
light_df
```

```python
light_df['시군구'] = light_df['도시']+' '+light_df['구']+' '+light_df['동']
```

```python
light_df
```

```python
train = pd.read_csv("/content/drive/MyDrive/data/dacon/train.csv")
```

```python
train
```

```python
train['시군구'].value_counts()
```

```python
merge_df = pd.merge(train,light,how = "left",on = '시군구')
```

```python
merge_df.info()
```

```python
merge_df['설치개수'].isnull().sum()
```

```python
train_check_address = []
for address in train['시군구'].value_counts().index.to_list():
  if address not in light['시군구'].value_counts().index.to_list():
    train_check_address.append(address)

print(train_check_address)
```

```python
train_check_address.sort()
```
